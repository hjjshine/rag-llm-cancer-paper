```python
soup = BeautifulSoup(requests.get('https://www.mycancergenome.org/content/drugs/pembrolizumab/').text, 'html.parser')
entries = soup.find_all('div', class_='about-drug-therapy-row')
len(entries)
```

```python
for entry in entries[-1:]:
    cancer = entry.find(class_='about-drug-therapy-header').text
    sensitivity = entry.find(class_='about-drug-therapy-sensitivity-row').text
    biomarker = entry.find(class_='small-12 columns biomarker-criteria').text
    clinical = entry.find(class_='small-12 columns response-setting-note').text
    print(cancer, sensitivity, biomarker, clinical)
```
